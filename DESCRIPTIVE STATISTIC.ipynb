### STACKED BAR





```python
dataa=pd.read_csv(r"C:\Users\simay\Desktop\Migrants.csv")
dataa.head(10)
#dataa["Region_of_Incident"].nunique
```

```python
import numpy as np
veri=dataa.loc[dataa["Region_of_Incident"].isin(["US-Mexico Border","Mediterranean","Middle East","North Africa",
                                    "Horn of Africa","Sub-Saharan Africa"])]

Summary=pd.crosstab(veri["Region_of_Incident"],veri["Reported_Year"],
                    values=veri["Total_Dead_and_Missing"],aggfunc=np.sum)
Summary

```

```python
flatui=["#66c2a5","#fc8d62","#8da0cb","#e78ac3","#a6d854","#ffd92f","#e5c494","#b3b3b3"]
plt.rcParams["figure.figsize"]=[10,8]

sns.set_context("notebook",font_scale=1.8,rc={"font_size":18,"axes.titlesize":20,"axes.labelsize":16})
sns.set_style("darkgrid")
Summary.plot(kind="bar",stacked=True,color=flatui)
plt.xticks(rotation=-45)
plt.title("Number of Missing Migrants \n Period of 2014-2017")
plt.ylabel("Number of deaths or  Missing Migrants")
plt.xlabel("Region of Causalities")
plt.show()








```

```python
#stacked=FALSE
flatui=["#66c2a5","#fc8d62","#8da0cb","#e78ac3","#a6d854","#ffd92f","#e5c494","#b3b3b3"]
plt.rcParams["figure.figsize"]=[10,8]

sns.set_context("notebook",font_scale=1.8,rc={"font_size":18,"axes.titlesize":20,"axes.labelsize":16})
sns.set_style("darkgrid")
Summary.plot(kind="bar",stacked=False,color=flatui)
plt.xticks(rotation=-45)
plt.title("Number of Missing Migrants \n Period of 2014-2017")
plt.ylabel("Number of deaths or  Missing Migrants")
plt.xlabel("Region of Causalities")
plt.show()





```

### MEASURES OF CENTER OF DATA



###### SAMPLE MEAN





```python
Mean=df_rate["Literacy_Rate"].mean()
print("The mean is ",round(Mean,2))


```

##### SAMPLE MEDIAN

```python
Median=df_rate["Literacy_Rate"].median()
print("The median is ",round(Median,2))



```

##### MEASURES OF SPREAD OR VARIATION OF DATA

##### MEAN ABSOLUTE DEVIATION 




```python
MAD=df_rate["Literacy_Rate"].mad()
print("The mean absolute deviation is :",round(MAD,3))
```

#### SAMPLE VARIANCE

```python
#compute sample standart deviation (variance) with a default ddof=1
var=df_rate["Literacy_Rate"].var()
print("Sample variance:",round(var,3))



```

#### POPULATION VARIANCE 

```python
#compute sample standart deviation (variance) with a default ddof=0
Pvar=df_rate["Literacy_Rate"].var(ddof=0)
print("Population  variance:",round(var,3))






```

#### SAMPLE STANDART DEVIATION 

```python
std=df_rate["Literacy_Rate"].std()
print("Sample standart devşatşon:",std)
```

#### POPULATION STANDART DEVIATION

```python
Pstd=df_rate["Literacy_Rate"].std(ddof=0)
print("Sample standart devşation:",round(Pstd,3))
```
